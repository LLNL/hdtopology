```python
import numpy as np
data = np.genfromtxt(fname='terrain1.csv', delimiter=',', skip_header = 1)
```

```python
# data
import matplotlib.pyplot as plt
plt.scatter(data[:,0], data[:,1], c=data[:,2])
```

```python
sampler = TDA_sampler()
```

```python
'''
    data is an numpy array 
    dimNames are the labels for the column of the array
'''
sampler.compute_TDA(data, dimNames=['D1','D2','f'])
```

```python
samples = sampler.sampling(extremaCount=3, samplePerExtrema=100)
```
